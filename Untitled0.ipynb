## 24. Разделимые наименьшие квадраты

[Каханер–Моулер–Нэш] 

Пусть необходимо аппроксимировать $m$ точек данных $(t_i,\, y_i)$ $(i=1,2,\dots,m)$ с помощью функции
$$
   b(t) = c_1 + c_2 t + c_3 t^2 + c_4 e^{\lambda t}.
$$
Параметры $c_1,\dots,c_4$ входят линейно, а $\lambda$ нелинейно.
Рассмотрим матрицу
$$
A(\lambda) =
\left(
\begin{array}{cccc}
    1 & t_1 & t_1^2 & e^{\lambda t_1}  \\
    1 & t_2 & t_2^2 & e^{\lambda t_2}  \\
    \vdots & \vdots & \vdots & \vdots  \\
    1 & t_m & t_m^2 & e^{\lambda t_m}  \\
\end{array}
\right)
$$
и векторы
$$
y = 
\left(
\begin{array}{c}
   y_1 \\ y_2 \\ \vdots \\ y_m
\end{array}
\right),
\qquad
c = 
\left(
\begin{array}{c}
   c_1 \\ c_2 \\ c_3\\ c_4
\end{array}
\right).
$$
Тогда задачу наименьших квадратов можно поставить следующим образом:
$$
\min_{\lambda}  \min_{c}  \|A(\lambda) c - y\|_2.
$$
Для любого заданного $\lambda$ внутренний минимум можно найти линейным методом наименьших квадратов.
Для нахождения внешнего минимума можно воспользоваться методом одномерной минимизации.

В таблице ниже приведены две группы данных, к которым необходимо применить описанный подход.
Единственным ли образом определяются неизвестные параметры?
Изобразите данные на графике и постройте графики построенных моделей.

```
       t        y         y
             Группа 1  Группа 2
 -----------------------------------
     0.00     20.00     20.00
     0.25     51.58     24.13
     0.50     68.73     26.50
     0.75     75.46     27.13
     1.00     74.36     26.00
     1.25     67.09     23.13
     1.50     54.73     18.50
     1.75     37.98     12.13
     2.00     17.28      4.00
```

````python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
m = 9
test = np.arange(0, 2.01, 0.25, dtype = float)

y_1 = np.array([20.00,51.58,68.73,75.46,74.36,67.09,54.73,37.98,17.28,])

y_2 = np.array([20.00,24.13,26.50,27.13,26.00,23.13,18.50,12.13,4.00,])
````

Функции для метода

````python
def f(lmd, C, x):
    return C[0] + C[1] * x + C[2] * (x ** 2) + C[3] * (np.e ** (lmd * x))

#Метод наименьших квадратов
def get(lmd, y):
    A =  np.ones((m, 4), dtype=float)
    A[:,1] = test
    A[:,2] = test ** 2
    A[:,3] = np.e ** (lmd * test)
    return np.linalg.lstsq(A, y)[0]

#Вычисление функции, которую надо минимизировать
def solve_1(lmd):    
    C = get(lmd, y_1)
    y_approx = f(lmd, C, test)
    return np.sum((y_approx - y_1) ** 2)
def solve_2(lmd):    
    C = get(lmd, y_2)
    y_approx = f(lmd, C, test)
    return np.sum((y_approx - y_2) ** 2)
````

Нанесём данные на график


````python
plt.grid()
plt.plot(test, y_1, 's')
plt.plot(test, y_2, 's')
````

Построим график функции 1, которую надо минимизировать


````python
x_a = np.arange(-10.0, 10.0, .01, dtype = float)
y_a = []
for x in x_a:
    y_a.append(solve_1(x))
    
plt.grid()
plt.plot(x_a, y_a)
````

Построим аппроксимацию

````python
lmd = minimize_scalar(solve_1, bounds=[-10, 10], method='bounded')['x']

x_v1 = np.arange(0, 2.01, 0.01, dtype = float)
C_result = get(lmd, y_1)
y_v1 = f(lmd, C_result, x_v1)
plt.grid()
plt.plot(x_v1, y_v1)
plt.plot(test, y_1, 's')
````

Построим график функции 2, которую надо минимизировать

````python
x_a = np.arange(-50, 10, .1, dtype = float)
y_a = []
for x in x_a:
    y_a.append(solve_2(x))
    
plt.plot(x_a, y_a)
````

Построим сразу обе выборки.

Функция возрастает идя к нулю, тогда пусть лямбда будет равна -50


````python
lmd = -50
x_v2 = np.arange(0, 2.01, 0.01, dtype = float)
C_result = get(lmd, y_2)
y_v = f(lmd, C_result, x_v1)

plt.grid()
plt.plot(x_v1, y_v1)
plt.plot(test, y_1, 's')
plt.plot(x_v2, y_v)
plt.plot(test, y_2, 's')
````

В первой функции минимум можно определить одним способом(внешняя функция имеет определённый минимум), а во второй это сделать не получится(она возрастает на этом промежутке), поэтому мы нашли приближение.
